```python
import numpy as np
np.random.seed(5)
A = np.random.randint(0,100,size=(9,9))
print("Elements in matrix")
print(A)
B=[]
```

```python
for i in range(0,A.shape[0],1):
  temp= A[0:].mean()
  B.append(temp)
print(B)
```

```python
C=np.zeros((A.shape[0],A.shape[1]))
zerosCount=0
nonZerosCount=0
for i in range(0,A.shape[0]):
  for j in range(0,A.shape[1]):
    if A[i,j] > B[i]:
      C[i,j]=0
      zerosCount=zerosCount+1
    else :
      C[i,j]=A[i,j]
      nonZerosCount=nonZerosCount+1


```

```python
if zerosCount+nonZerosCount==A.size:
  print("Updated matrix is validated")

else:
  print("Non validated")

print("Updated matrix")
print(C)
print("The numbner of zeros are:",zerosCount)
print("The numbner of Non-zeros are:",nonZerosCount)
```
